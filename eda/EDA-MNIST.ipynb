```python
import math

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mnist_data = np.load("../dataset/mnist_test_seq.npy")
mnist_data.shape
```

```python
letter_types = mnist_data[:, 1, :, :]

col_count = 4
fig, ax = plt.subplots(math.ceil(len(letter_types) / col_count), col_count, figsize=(10, 20))

for i in range(len(letter_types)):
    row_idx, col_idx = i // col_count, i % col_count
    
    ax[row_idx, col_idx].imshow(letter_types[i])


plt.show()
```
